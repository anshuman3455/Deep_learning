```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\user\\Desktop\\data sets\\Attrition.csv")
```

```python
df.head()
```

```python
import tensorflow as tf 
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = df.dropna()
```

```python
x = df.drop(columns = 'Attrition', axis = 1)
y = df["Attrition"]
```

```python
x.info()
```

```python
#Preprocessing categorical data
categroical_features = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18',
                       'OverTime']
```

```python
x_encoded = pd.get_dummies(x,columns = categroical_features, drop_first = True )
```

```python
# Encode target column
lb = LabelEncoder()
y_encoded = lb.fit_transform(y)
y_encoded
```

```python
x_train,x_test, y_train,y_test = train_test_split(x_encoded,y_encoded,test_size=0.2 , random_state=9)
```

```python
sc = StandardScaler()
```

```python
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

# Now we will create architexture of our model

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 128 , activation = 'relu', input_dim = x_train_scaled.shape[1]),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units = 64 , activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units = 1 , activation='sigmoid')
])
```

```python
# Define loss function and metris 
loss_fn = tf.keras.losses.BinaryCrossentropy()
metrics = ['accuracy']
```

```python
# Define loss function and metrics 
learning_rate = 0.001
mometum = 0.9
optimizer = tf.keras.optimizers.SGD(learning_rate = learning_rate, momentum=mometum, nesterov=True)
```

```python
# Compile the model
model.compile(optimizer= optimizer, loss =loss_fn , metrics = metrics)
model.fit(x_train_scaled, y_train, epochs = 50 , batch_size = 8 , validation_split= 0.1)
```

```python
# Evaluate the model on test set 
loss ,accuracy = model.evaluate(x_test_scaled , y_test)
print(f"Test loss ={loss:4f}, Test Accuracy = {accuracy : 4f}")
```

```python
predictions = model.predict(x_test_scaled)
predictions
```
