# Zero weights 
In this dataset We will initialize the weights and baises to zero. Purpose of doing this to cheak if our weights is zero then what will be the model of deep learning will perform like

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=100,noise=0.175)
```

```python
plt.scatter(x=x[:,0],y=x[:,1],c=y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(25,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
# To see weight which was set by keras using (deafult=Xavier_uniform)
model.get_weights()
```

```python
zero_weights = model.get_weights()
```

```python
zero_weights[0] = np.zeros(model.get_weights()[0].shape)
zero_weights[1] = np.zeros(model.get_weights()[1].shape)
zero_weights[2] = np.zeros(model.get_weights()[2].shape)
zero_weights[3] = np.zeros(model.get_weights()[3].shape)
```

```python
# initializing weights to zero 
model.set_weights(zero_weights)
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history= model.fit(x,y,epochs = 100, validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y,clf=model, legend=2)
```

## Relu and tanh activation function
The both activation function weight will not do any updation in their respective weights this brings that no use of backproption in the model. # zero weights
## Sigmoid activation function
This activation function will do update in its weighs but all layer of the model behave like a single perceptorn this make them a single perceptron in one layer. (sigmoid weight =0 means =0.5(because of non zero centroid activaiton function))

# Same weights to all neuron
In this all weights will be initialized as a same value 

```python
model1 = Sequential()

model1.add(Dense(2,activation='sigmoid',input_dim=2))
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

```python
model1.get_weights()
```

```python
initial_weights = model1.get_weights()

initial_weights[0] = np.ones(model.get_weights()[0].shape)*0.5
initial_weights[1] = np.ones(model.get_weights()[1].shape)*0.5
initial_weights[2] = np.ones(model.get_weights()[2].shape)*0.5
initial_weights[3] = np.ones(model.get_weights()[3].shape)*0.5
model1.set_weights(initial_weights)

model1.get_weights()
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history = model1.fit(x,y,epochs=100,validation_split=0.2)
```

```python
model1.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y, clf=model1, legend=2)
```
